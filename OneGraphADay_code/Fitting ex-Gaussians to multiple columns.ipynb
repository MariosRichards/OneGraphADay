```python
import numpy as np
from scipy.stats import exponnorm
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# dummy dataframe of 4 columns by 100 rows
input_df = pd.DataFrame(np.random.randn(100,4))
input_df
```

```python
num_cols = input_df.shape[1]

for col in range(0,num_cols)

fig, ax = plt.subplots(2, num_cols, sharex=True)


params = exponnorm.fit(input_df[col])
print(params)
K = params[0]

x = np.linspace(exponnorm.ppf(0.01, K),\
                exponnorm.ppf(0.99, K), 100)
ax[0].plot(x, exponnorm.pdf(x, params[0],params[1],params[2]),\
        'r-', lw=5, alpha=0.6, label='exponnorm pdf')

input_df[col].hist(ax=ax[0],bins=20)
```

```python
num_cols = input_df.shape[1]

fig, ax = plt.subplots(1, input_df.shape[1],sharex=True, figsize = (20,5))

x = np.linspace(-4,4,100)

bins = 20

for col in range(0,num_cols):

    params = exponnorm.fit(input_df[col])
    print(params)



    input_df[col].hist(ax=ax[col], bins=bins)

    eg_dist = exponnorm.pdf(x, params[0],params[1],params[2])
    
    # edges of distribution lead to overflows -> leads to nans -> lead to plotting failing
    # so, replace with 0s
    eg_dist = np.nan_to_num(eg_dist)    

    adj_factor = (input_df.shape[0]/bins)/np.mean( eg_dist )

    ax[col].plot(x, exponnorm.pdf(x, params[0],params[1],params[2])*adj_factor,\
            'r-', lw=5, alpha=0.6, label='exponnorm pdf')

```

```python
# The model function, f(x, ...).
# It must take the independent variable as the first argument and the parameters to fit as separate remaining arguments.

def curve_fit_exg(x, size_fit, var1, var2, var3):
    return size_fit * exponnorm.pdf(x, var1, var2, var3)
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# def func(x, a, b, c):
#     return a * np.exp(-b * x) + c

# define the data to be fit with some noise

data = pre_df_mean

xdata = data["BMI"]
# y = func(xdata, 2.5, 1.3, 0.5)
# y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = data["mean_attractiveness"]-1
plt.plot(xdata, ydata, 'b-', label='data')

# Fit for the parameters a, b, c of the function `func`

popt, pcov = curve_fit(curve_fit_exg, xdata, ydata)
plt.plot(xdata, curve_fit_exg(xdata, *popt), 'r-', label='fit')

perr = np.sqrt(np.diag(pcov))
print( popt, perr )

# # Constrain the optimization to the region of ``0 < a < 3``, ``0 < b < 2``
# # and ``0 < c < 1``:

# popt, pcov = curve_fit(func, xdata, ydata, bounds=(0, [3., 2., 1.]))
# plt.plot(xdata, func(xdata, *popt), 'g--', label='fit-with-bounds')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
