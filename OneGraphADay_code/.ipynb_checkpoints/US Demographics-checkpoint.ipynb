```python
# in case we need to do any stats
import statsmodels.api as sm
# ignore warning message - I'm sure nothing will go wrong
```

```python
# graphics modules
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
query = ("https://data.cdc.gov/resource/pzzm-ur7y.json")
raw_data = pd.read_json(query)
```

```python
# top 10 rows of the dataset
raw_data.head(10)
```

```python
# How they split people into 'races'
raw_data["race"].unique()
```

```python
# Quick breadown of the dataset columns
```

```python
# birth_rate = annual number of births / 1000 population
# fertility_rate = annual number of births / 1000 women age 15–44 or 15–49
# (checked, I believe US uses 15-49 as 'childbearing years' - I assume inclusive)
```

```python
# I have no particular sense of what an annual fertility rate of 60/1000 is
# but if we divide through by the number of childbearing years we get the Total Fertility Rate
# i.e. average lifetime number of children born to each woman (if the trend in that year continued for their whole lives)
# Which seems salient to me (because you know replacement is ~2)

raw_data["Total Fertility Rate"] = raw_data["fertility_rate"] / (49-15+1) # because it's inclusive
```

```python
# key difference - fertility_rate won't change if your population structure changes, but individual fertility rates stay same
# people quoting birth_rates when talking about fertility? Either (i) honest ignorance (ii) deliberate shysterism
# (because differences that are due to an ageing population will seem to be due to differential fertility)
```

```python
# live_births = pretty self-explanatory (I note that the big spikes correspond to big changes in US abortion law
# Roe vs Wade started in 1971, next big redefinition in 1992
# Oddly, it looks almost as if the the legal challenges were *driven* by the birth spikes)
```

```python
# Race
# categories here - 'Asian/Pacific Islander', 'American Indian/Alaska Native', 'White', 'Black'
# fairly limited set of categories, but that's race for you - the distinctions people stress about 40 years ago may
# be quite different from the ones they stress about now. When people started collecting data in 1960, they might not have
# seen any point in distinguishing between "latino White" and "non-latino White" - in fact, there may well have been no
# difference in birth rates to justify the distinction
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.tsplot(raw_data, ax=ax, time="year", unit="race",condition ="race", value="Total Fertility Rate");

```

```python
# these graphs included so you can see how useless they are
fig, ax = plt.subplots(1,2, figsize=(20,10))

sns.tsplot(raw_data, ax=ax[0],time="year", unit="race",condition ="race", value="birth_rate" );
sns.tsplot(raw_data, ax=ax[1], time="year", unit="race",condition ="race", value="live_births" );
```

```python
# PROJECTION

# I did play around fitting curves to the raw data - but I think two things are obvious from looking at the data
# (1) All groups are following the trends of falling fertility rates
# (2) Almost all groups are converging on the same rate (just below replacement - not just US, general trend of pretty much everywhere)
# (3) Political/economic events that act on racial groups *act on all racial groups*
# (4) The differences are about where groups are in their position on the transition from high fertility to low
# (most probably due to economic differences)
# (5) The only obvious outlier group is American Indian/Alaska Natives - their fertility rate is just dropping off
# the bottom of the scale (of course, that might simply represent their extreme economic circumstances,
# relative to other group's distributions)
```

```python
# CONCLUSION
# Projected fertiliy rates of all of these US racial groups *other than American Indian/Alaska Natives*
# Is that it appears to be a flat line (to which all groups have converged, albeit at different speed)
# The line is at TFR 1.8 with a lot of noises/oscillation
# But that's enough to get a sense of what the consequences are for the future population
# Most particularly the age structure

# Oh - and something very bad is happening with American Indian/Alaska Natives
```
